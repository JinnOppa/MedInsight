```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path = "C:\\Users\\Republic Of Gamers\\OneDrive\\Documents\\GitHub\\TSDN-BoyWithLuv\\Source\\Data\\Patients Data ( Used for Heart Disease Prediction ).xlsx"
df = pd.read_excel(path)
df
```

```python
columns = ['HadAngina', 'HadHeartAttack', 'HadStroke', 'HadSkinCancer', 'HadDepressiveDisorder', 'HadArthritis']

fig, axes = plt.subplots(3, 3, figsize=(15, 15))
axes = axes.flatten()

for i, col in enumerate(columns):
    ax = sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(col)
    axes[i].set_xlabel("Count")
    axes[i].set_ylabel("Frequency")
    
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.text(
                x=p.get_x() + p.get_width() / 2,
                y=height,
                s=f'{int(height)}',
                ha='center',
                va='bottom',
                fontsize=10,
                color='black'
            )

# Remove unused subplots
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
columns = ['HadAngina', 'HadHeartAttack', 'HadStroke', 'HadSkinCancer', 'HadDepressiveDisorder', 'HadArthritis']

fig, axes = plt.subplots(3, 3, figsize=(15, 15))
axes = axes.flatten()

for i, col in enumerate(columns):
    # Calculate counts for each category
    counts = df[col].value_counts()
    labels = counts.index
    sizes = counts.values
    
    # Plot the pie chart
    axes[i].pie(
        sizes, 
        labels=labels, 
        autopct='%1.1f%%', 
        startangle=90, 
        textprops={'fontsize': 10}
    )
    axes[i].set_title(col)

# Remove unused subplots
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
