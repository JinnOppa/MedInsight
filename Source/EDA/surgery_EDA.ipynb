```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path = "C:\\Users\\Republic Of Gamers\\OneDrive\\Documents\\GitHub\\TSDN-BoyWithLuv\\Source\\Data\\surgery_record.csv"
df = pd.read_csv(path)
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12,7))
sns.countplot(data = df, x = 'preoperative_risk_class')
plt.xlabel("Risks")
plt.ylabel("Count")
plt.title("Pre-Operative Risk Class Distribution")
plt.show()
```

```python
# Calculate the counts for each risk class
risk_counts = df['preoperative_risk_class'].value_counts()

# Create a blue gradient color palette
colors = plt.cm.Blues(np.linspace(0.4, 0.9, len(risk_counts)))

# Create a pie chart
fig, ax = plt.subplots(figsize=(12, 7))
ax.pie(
    risk_counts, 
    labels=risk_counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=colors
)

# Set the title
ax.set_title("Pre-Operative Risk Class Distribution")
plt.show()
```
