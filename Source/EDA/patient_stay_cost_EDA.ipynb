```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path = "C:\\Users\\Republic Of Gamers\\OneDrive\\Documents\\GitHub\\TSDN-BoyWithLuv\\Source\\Data\\patient_cost_and_stay_data.csv"
df = pd.read_csv(path)
df
```

```python
fig, ax = plt.subplots(figsize = (12,7))
sns.countplot(data = df, x = 'stay_duration')
plt.xlabel("Stay Duration")
plt.ylabel("Count")
plt.title("Patient Stay Duration Distribution")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,7))
sns.histplot(data = df, x = 'total_cost')
plt.xlabel("Hospital Total Cost")
plt.ylabel("Count")
plt.title("Total Hospital Cost Distribution")
plt.show()
```
