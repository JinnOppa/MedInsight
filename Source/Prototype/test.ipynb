```python
df = df.drop(['Name', 'Doctor', 'Hospital','Room Number','Discharge Date','Date of Admission'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode_gender = LabelEncoder()
encode_blood_type = LabelEncoder()
encode_medical_condition = LabelEncoder()
encode_insurance_provider = LabelEncoder()
encode_admission_type = LabelEncoder()
encode_medication = LabelEncoder()
encode_test_result = LabelEncoder()
```

```python
df['Gender'] = encode_gender.fit_transform(df['Gender'])
df['Blood Type'] = encode_blood_type.fit_transform(df['Blood Type'])
df['Medical Condition'] = encode_medical_condition.fit_transform(df['Medical Condition'])
df['Insurance Provider'] = encode_insurance_provider.fit_transform(df['Insurance Provider'])
df['Admission Type'] = encode_admission_type.fit_transform(df['Admission Type'])
df['Medication'] = encode_medication.fit_transform(df['Medication'])
df['Test Results']= encode_test_result.fit_transform(df['Test Results'])
```

```python
df.info()
```

```python
df.corr()
```

```python
df_sample = df.sample(n=10000, random_state=42)
df_sample
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_sample['Billing Amount'].quantile(0.25)
Q3 = df_sample['Billing Amount'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_no_outliers = df_sample[(df_sample['Billing Amount'] >= lower_bound) & (df_sample['Billing Amount'] <= upper_bound)]

# Check the shape of the new DataFrame
df_no_outliers.shape
```

```python
x = df_no_outliers.drop('Billing Amount', axis=1)
y = df_no_outliers['Billing Amount']
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rf_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

xgb_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.5, 0.7, 1.0],
    'colsample_bytree': [0.5, 0.7, 1.0]
}
```

```python
rf_model = RandomForestRegressor(random_state=42) 
```

```python
rf_grid_search = GridSearchCV(estimator=rf_model, param_grid=rf_param_grid,
                               scoring='neg_mean_squared_error', cv=3, verbose=1, n_jobs=-1)
rf_grid_search.fit(x_train, y_train)
```

```python
print("Best parameters for Random Forest:", rf_grid_search.best_params_) 
```

```python
rf_best_model = rf_grid_search.best_estimator_
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

rf_predictions = rf_best_model.predict(x_test)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_predictions))
rf_r2 = r2_score(y_test, rf_predictions)
print(f"Random Forest RMSE: {rf_rmse:.2f}, R²: {rf_r2:.2f}")
```
