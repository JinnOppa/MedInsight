```python
def create_admission_data(n_samples):
    np.random.seed(42)
    
    # Generate Patient Demographics
    data = pd.DataFrame({
        'patient_id': np.arange(1, n_samples + 1),
        'gender': np.random.choice(['Male', 'Female'], n_samples, p=[0.48, 0.52]),
        'age': np.random.normal(50, 18, n_samples).clip(18, 90).round().astype(int),
        'insurance_type': np.random.choice(['Private', 'Self-Pay', 'Social Security Agency'], n_samples, p=[0.5, 0.05, 0.45])
    })
    
    # Generate Patient Lifestyle factors with gender and age influence
    data['smoking_status'] = np.where(
        data['gender'] == 'Female', 
        np.where(data['age'] < 21, 'Never', np.random.choice(['Never', 'Former', 'Current'], n_samples, p=[0.65, 0.25, 0.1])),
        np.where(data['age'] < 21, 'Never', np.random.choice(['Never', 'Former', 'Current'], n_samples, p=[0.6, 0.25, 0.15]))
    )
    data['e_cigarette_usage'] = np.where(
        data['gender'] == 'Female', 
        np.where(data['age'] < 21, False, np.random.choice([True, False], n_samples, p=[0.12, 0.88])),
        np.where(data['age'] < 21, False, np.random.choice([True, False], n_samples, p=[0.1, 0.9]))
    )
    data['alcohol_consumption_rate'] = np.where(
        data['gender'] == 'Female', 
        np.where(data['age'] > 65, 'None', np.random.choice(['None', 'Occasional', 'Moderate', 'Heavy'], n_samples, p=[0.25, 0.55, 0.18, 0.02])),
        np.where(data['age'] > 65, 'None', np.random.choice(['None', 'Occasional', 'Moderate', 'Heavy'], n_samples, p=[0.2, 0.5, 0.25, 0.05]))
    )
    
    # Generate Medical History with more surgery options
    surgery_names = ['Appendectomy', 'Gallbladder Removal', 'Knee Replacement', 'Heart Bypass', 
                     'Cataract Surgery', 'Hip Replacement', 'Hernia Repair', 'Spinal Fusion', 
                     'Breast Cancer Surgery', 'Liver Transplant']
    data['previous_admission_count'] = np.random.poisson(1, n_samples).clip(0, 5)
    data['surgery_name'] = np.random.choice(surgery_names, n_samples)
    data['surgery_type'] = np.where(data['surgery_name'].isin(['Heart Bypass', 'Knee Replacement', 'Hip Replacement', 'Spinal Fusion', 'Breast Cancer Surgery', 'Liver Transplant']), 'Major', 'Minor')
    data['surgery_duration'] = np.where(data['surgery_type'] == 'Major', 
                                        np.random.normal(180, 30, n_samples).clip(60, 300), 
                                        np.random.normal(60, 20, n_samples).clip(20, 120)).round().astype(int)

    # Generate Patient Hospital Stay details with room type restrictions based on insurance
    data['room_type'] = np.where(
        (data['insurance_type'] == 'Social Security Agency'),
        np.random.choice(['ICU', 'Regular Ward'], n_samples, p=[0.7, 0.3]),
        np.where(data['surgery_type'] == 'Major', 
                 np.random.choice(['ICU', 'Private Ward', 'VIP Ward'], n_samples, p=[0.5, 0.3, 0.2]), 
                 np.random.choice(['Regular Ward', 'Private Ward'], n_samples, p=[0.8, 0.2]))
    )
    data['medical_equipment_count'] = np.where(data['surgery_type'] == 'Major', 
                                               np.random.poisson(2, n_samples).clip(1, 5), 
                                               np.random.poisson(1, n_samples).clip(0, 3))
    data['stay_duration'] = np.where(data['surgery_type'] == 'Major', 
                                     np.random.normal(10, 4, n_samples).clip(5, 30), 
                                     np.random.normal(3, 2, n_samples).clip(1, 10)).round().astype(int)

    # Define fixed ward costs based on room type
    ward_costs = {'ICU': 800.0, 'Regular Ward': 300.0, 'Private Ward': 500.0, 'VIP Ward': 1000.0}
    data['ward_cost'] = data['room_type'].map(ward_costs)

    # Generate Financial Details with float values and revised total cost formula
    data['surgery_cost'] = np.where(data['surgery_type'] == 'Major', 
                                    np.random.normal(15000, 3000, n_samples).clip(10000, 30000).round(2), 
                                    np.random.normal(5000, 1000, n_samples).clip(3000, 10000).round(2))
    
    # Adjust medication_cost based on medical_equipment_count
    base_medication_cost = np.where(data['surgery_type'] == 'Major', 
                                    np.random.normal(2000, 500, n_samples).clip(1000, 4000), 
                                    np.random.normal(500, 200, n_samples).clip(100, 1000))
    equipment_multiplier = 1 + (data['medical_equipment_count'] * 0.05)  # 5% increase per equipment item
    data['medication_cost'] = (base_medication_cost * equipment_multiplier).round(2)

    # Calculate the total cost
    data['total_cost'] = (data['surgery_cost'] + data['medication_cost'] + (data['stay_duration'] * data['ward_cost'])).round(2)

    return data
```

```python
df = create_admission_data(500000)
```
